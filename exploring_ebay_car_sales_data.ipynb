# Data Analysis: Exploring eBay Car Sales Data

In this project, we analyze a dataset of used car listings from *eBay Kleinanzeigen*, a classifieds section of the German eBay website.

### Our goals

Our primary goal is to **clean the data** and **perform an initial analysis** to understand the characteristics of the used car market in Germany.

### Dataset Description

The dataset includes various attributes such as the car's price, brand, registration year, mileage (odometer), and fuel type, providing a realistic look at scraped web data that requires significant cleaning.

## 1. Loading the Dataset

To begin our analysis, we will import the necessary libraries—**pandas** and **NumPy**—and load the `autos.csv` dataset. 

Because this dataset was scraped from the German eBay website, it contains special characters that aren't compatible with the standard `UTF-8` encoding. We will use the `latin1` encoding to ensure the data loads correctly.

```python
import pandas as pd
import numpy as np

# Attempt to read the CSV file
# We use 'latin1' encoding as the default 'UTF-8' may throw an error for this specific file
autos = pd.read_csv('autos.csv', encoding='latin1')

autos  # Display the first and last few rows of the dataframe
```

```python
# Display summary information and the first 5 rows
print(autos.info())
autos.head()
```

## 2. Initial Observations

After inspecting the data, several key observations can be made:

1. **Column Names**: The dataset uses CamelCase (e.g., `yearOfRegistration`) instead of Python's preferred snake_case (e.g., `year_of_registration`). Some column names could also be more descriptive.
2. **Missing Values**: Columns such as `vehicleType`, `gearbox`, `model`, `fuelType`, and `notRepairedDamage` contain null values.
3. **Data Types**: 
    * The `price` and `odometer` columns are stored as text (strings) because they likely include characters like `$` or `km`. We will need to clean these and convert them to numeric types.
    * Dates (`dateCrawled`, `dateCreated`, `lastSeen`) are currently strings and will need to be converted to actual datetime objects for analysis.
4. **General Content**: Some columns like `seller` and `offerType` appear to have mostly identical values, which might make them less useful for analysis.

## 3. Cleaning Column Names

In this step, we will clean the column headers to make the dataset easier to work with in Python. The changes will include:

1.  **Converting from camelCase to snake_case**: Following Python's PEP 8 style guide, we will replace names like `dateCrawled` with `date_crawled`. This makes the code more readable and consistent.
2.  **Renaming for Clarity**: Based on the data dictionary, we will rename several columns to be more descriptive:
    * `yearOfRegistration` → `registration_year`
    * `monthOfRegistration` → `registration_month`
    * `notRepairedDamage` → `unrepaired_damage`
    * `dateCreated` → `ad_created`

By standardizing the headers now, we prevent errors in future steps and make it easier to reference columns using dot notation or bracket indexing.

```python
# Print the existing column names
print(autos.columns)

# Create a modified list of column names
new_columns = [
    'date_crawled', 'name', 'seller', 'offer_type', 'price', 'abtest',
    'vehicle_type', 'registration_year', 'gearbox', 'power_ps', 'model',
    'odometer', 'registration_month', 'fuel_type', 'brand',
    'unrepaired_damage', 'ad_created', 'nr_of_pictures', 'postal_code',
    'last_seen'
]

# Assign the new names back to the dataframe
autos.columns = new_columns

# Check the results
autos.head()
```

## 4. Initial Exploration and Cleaning

Before proceeding with deep analysis, we use `df.describe()` to get an overview of the dataset. We are looking for:
* **Static Columns**: Features where almost all values are the same (low variance), which typically don't provide useful insights.
* **Hidden Numeric Data**: Numeric values currently stored as text because of symbols like `$` or `km`.
* **Anomalies**: Unusual distributions, such as columns that appear to be empty or have invalid data.

```python
# Explore descriptive statistics for both numeric and categorical columns
autos.describe(include='all')
```

### Observations and Cleaning Plan

Based on the summary above, we can conclude the following:

1.  **Drop Candidates**: The `seller` and `offer_type` columns contain nearly 100% identical values. Additionally, `nr_of_pictures` seems to contain only zeros. We will investigate these further.
2.  **Numeric Conversion**: The `price` and `odometer` columns are currently stored as `object` (text) due to the presence of symbols ($ and km) and commas.

**Cleaning Steps:**
* Remove the `$` and `,` characters from the `price` column and convert it to an integer.
* Remove the `km` and `,` characters from the `odometer` column, convert it to an integer, and rename it to `odometer_km` to preserve the unit information.

```python
# Cleaning the 'price' column
autos["price"] = (autos["price"]
                  .str.replace("$", "")
                  .str.replace(",", "")
                  .astype(int)
                 )

# Cleaning the 'odometer' column
autos["odometer"] = (autos["odometer"]
                     .str.replace("km", "")
                     .str.replace(",", "")
                     .astype(int)
                    )

# Rename 'odometer' to 'odometer_km'
autos.rename({"odometer": "odometer_km"}, axis=1, inplace=True)

# Verify the result
autos[["price", "odometer_km"]].head()
```

## 5. Exploring the Odometer and Price Columns

Now that we have converted the `price` and `odometer_km` columns to numeric types, we will examine them for **outliers**—values that are unrealistically high or low. These anomalies often occur due to data entry errors or issues with the original scraping process.

We will analyze the distributions and use the `df.between()` method to remove any data points that could skew our analysis.

```python
# Exploring Odometer
print("Odometer Unique Values:", autos["odometer_km"].unique().shape[0])
print(autos["odometer_km"].describe())
print("\nOdometer Value Counts (Sorted):")
print(autos["odometer_km"].value_counts().sort_index(ascending=False).head())

# Exploring Price
print("\nPrice Unique Values:", autos["price"].unique().shape[0])
print(autos["price"].describe())
print("\nTop 10 Highest Prices:")
print(autos["price"].value_counts().sort_index(ascending=False).head(10))
print("\nTop 10 Lowest Prices:")
print(autos["price"].value_counts().sort_index(ascending=True).head(10))
```

### Observations and Outlier Removal

**Odometer:**
The odometer values appear to be rounded, suggesting users chose from a pre-set list of options. The maximum value is 150,000 km and the minimum is 5,000 km. There are no obvious outliers here that need removal.

**Price:**
The price column, however, has several issues:
* There are **1,421** listings with a price of **\$0**. Since eBay is an auction/sales site, \$0 is not a realistic starting price and these rows should be removed.
* There are extreme high-end outliers, such as listings for **\$99,999,999** and **\$12,345,678**. 
* While some luxury cars (like Ferraris) can legitimately sell for high amounts, prices jump significantly after **\$350,000**. 

We will keep listings between **\$1** and **\$350,000**, as anything above that tends toward unrealistic "placeholder" numbers.

```python
# Removing outliers in price
autos = autos[autos["price"].between(1, 350000)]

# Verify the new distribution
print(autos["price"].describe())
```

## 7. Dealing with Incorrect Registration Year Data

As noted in the previous step, the `registration_year` column contains several inaccuracies that will bias our analysis. 

**Deciding the Bounds:**
* **Upper Bound**: Since the data was crawled in **2016**, any registration year above 2016 is physically impossible (a car cannot be registered after the ad was already seen).
* **Lower Bound**: While the first cars were invented in the late 1800s, it is highly unlikely that a significant number of listings on eBay Kleinanzeigen are for cars from that era. We will set the lower bound at **1900**.

Let's check what percentage of our data falls outside the **1900–2016** range to determine if we can safely remove them.

```python
# Check how many rows are outside the 1900-2016 range
out_of_range = autos[~autos["registration_year"].between(1900, 2016)]
print(f"Number of invalid rows: {out_of_range.shape[0]}")
print(f"Percentage of invalid rows: {(out_of_range.shape[0] / autos.shape[0]) * 100:.2f}%")

# Filter the dataframe to keep only valid registration years
autos = autos[autos["registration_year"].between(1900, 2016)]

# Calculate the distribution of the remaining values
print("\nTop 10 Registration Years (Normalized):")
print(autos["registration_year"].value_counts(normalize=True).head(10))
```

### Observations on Registration Years

By removing the invalid years, we only lost about **3.9%** of the data, which is acceptable for maintaining the integrity of our analysis.

**Key Findings:**
* Most of the vehicles were registered within the last 20 years (specifically the **late 1990s to late 2000s**).
* The years **2000, 2005, and 1999** are the most common registration years in the dataset.
* This distribution aligns with what we would expect from a used car market, where the majority of listings are for relatively modern vehicles rather than vintage collectibles.

## 8. Exploring Price by Brand

Now that the data is clean, we can explore how different car brands compare in terms of pricing. We will focus our analysis on brands that represent a significant portion of the market. 

First, let's look at the market share of each brand.

```python
# Calculate the percentage of listings for each brand
brand_counts = autos["brand"].value_counts(normalize=True)
print(brand_counts)

# Select brands that represent more than 5% of the total listings
common_brands = brand_counts[brand_counts > 0.05].index
print("\nSelected Brands for Aggregation:", common_brands)
```

### Selection Strategy
The top 6 brands in the dataset all have a market share greater than 5%. **Volkswagen** is by far the most popular brand, nearly doubling the share of the next most frequent brands, **BMW** and **Opel**. 

We will limit our aggregation to these top 6 brands (Volkswagen, BMW, Opel, Mercedes Benz, Audi, and Ford) to ensure our mean prices are based on a large enough sample size to be representative.

```python
# Create an empty dictionary to store aggregate data
brand_mean_prices = {}

# Loop over the selected brands
for brand in common_brands:
    # Subset the dataframe for the specific brand
    brand_only = autos[autos["brand"] == brand]
    # Calculate the mean price
    mean_price = brand_only["price"].mean()
    # Store in dictionary
    brand_mean_prices[brand] = int(mean_price)

# Display the results
brand_mean_prices
```

### Analysis of Results
From the aggregated price data, we can observe a clear "price tier" among the top brands:

* **Luxury Tier**: **Audi**, **BMW**, and **Mercedes Benz** are the most expensive, with average prices between **\$8,000 and \$9,000**.
* **Mid Tier**: **Volkswagen** sits in the middle with an average price of around **\$5,400**. This likely explains its massive popularity, as it offers a balance of quality and affordability.
* **Budget Tier**: **Ford** and **Opel** are the most affordable options, with average prices between **\$3,000** and **\\$3,700**.

This suggests that even in the used car market, premium branding significantly influences the listing price.

## 9. Storing Aggregate Data in a DataFrame

While we have identified a price gap between luxury and budget brands, we don't yet know if this is driven by the condition of the cars (such as mileage) or purely by brand value. 

In this step, we will calculate the average mileage for the top 6 brands and combine both the mean price and mean mileage into a single DataFrame. This will allow us to easily compare the two metrics side-by-side.

```python
# Create an empty dictionary for mean mileage
brand_mean_mileage = {}

# Calculate mean mileage for the top brands
for brand in common_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_mileage = brand_only["odometer_km"].mean()
    brand_mean_mileage[brand] = int(mean_mileage)

# Convert both dictionaries to Series objects
mean_price_series = pd.Series(brand_mean_prices)
mean_mileage_series = pd.Series(brand_mean_mileage)

# Create a DataFrame from the mean_price_series
brand_info = pd.DataFrame(mean_price_series, columns=['mean_price'])

# Assign the mean_mileage_series as a new column
brand_info['mean_mileage'] = mean_mileage_series

# Display the combined data
brand_info
```

### Analysis of Price vs. Mileage

By comparing the average price and mileage across the top brands, we can draw the following conclusions:

* **Consistent Mileage**: Interestingly, the average mileage for all these brands is very similar, falling mostly between **128,000 km and 130,000 km**. 
* **Brand Value Over Condition**: Despite the luxury brands (Audi, BMW, and Mercedes-Benz) having similar or even slightly higher mileage than the budget brands (Opel and Ford), they still command a much higher average price.
* **The Premium Gap**: For example, **Audi** has a higher average mileage than **Opel**, yet its average price is nearly **3 times higher**.

This confirms that the price differences we observed earlier are primarily driven by brand prestige and original market positioning rather than the luxury cars being in "newer" condition than the budget ones.

## 9. Further Cleaning and Final Analysis

To conclude the project, we will perform two additional tasks:
1.  **Translating German Categories**: Several columns use German terms (e.g., `manuell` for manual). We will map these to English to make the dataset accessible to a wider audience.
2.  **Damage Analysis**: We will compare the average price of cars with unrepaired damage versus those that have been repaired or have no damage.

```python
# Check unique values in gearbox and fuel_type
print("Gearbox unique values:", autos["gearbox"].unique())
print("Fuel type unique values:", autos["fuel_type"].unique())

# Define mapping dictionaries
gearbox_map = {
    'manuell': 'manual',
    'automatik': 'automatic'
}

fuel_type_map = {
    'benzin': 'petrol',
    'diesel': 'diesel',
    'lpg': 'lpg',
    'pöl': 'vegetable_oil',
    'hybrid': 'hybrid',
    'elektro': 'electric',
    'andere': 'other'
}

# Apply mappings
autos["gearbox"] = autos["gearbox"].map(gearbox_map)
autos["fuel_type"] = autos["fuel_type"].map(fuel_type_map)

# Verify translation
autos[["gearbox", "fuel_type"]].head()
```

```python
# First, translate the values to English
damage_map = {
    'ja': 'yes',
    'nein': 'no'
}
autos["unrepaired_damage"] = autos["unrepaired_damage"].map(damage_map)

# Now run the aggregation
damage_impact = autos.groupby("unrepaired_damage")["price"].mean()

print("Average Price based on Damage Status:")
print(damage_impact)

price_diff = damage_impact['no'] - damage_impact['yes']
print(f"\nCars without damage are, on average, ${price_diff:.2f} more expensive.")
```
